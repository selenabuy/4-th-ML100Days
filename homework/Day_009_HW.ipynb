```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
```

```python
# 設定 data_path
dir_data = 'D:\\Source\\ML100Days_Data\\D4'

f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 秀出資料欄位的類型與數量
dtype_df = app_train.dtypes.reset_index()
dtype_df.columns = ["Count", "Column Type"]
dtype_df = dtype_df.groupby("Column Type").aggregate('count').reset_index()
dtype_df
```

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""

##怪怪的
#dtype_select = ['float64']
#numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])

numeric_columns = []

for dtype, feature in zip(app_train.dtypes, app_train.columns):
    #print(dtype,':',feature)
    if dtype == 'float64':
        numeric_columns.append(feature)
    elif dtype == 'int64':
        numeric_columns.append(feature)


        
# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
print("Numbers of remain columns %s" % (len(numeric_columns)))


# 檢視這些欄位的數值範圍
for col in numeric_columns:
    """
    Your CODE HERE, make the box plot
    """
    print(app_train[col].describe())
    plt.hist(app_train[col], bins=50, color='SteelBlue', edgecolor="LightSteelBlue")
    plt.show()
    sns.boxplot(x=app_train[col])
    plt.show()
    plt.plot(app_train[col],'*')
    plt.show()
```

```python
#可疑的欄位
#DAYS_EMPLOYED How many days before the application the person started current employment
#最大值掉在分配外面
print(app_train['DAYS_EMPLOYED'].describe())
plt.hist(app_train['DAYS_EMPLOYED'], bins=50, color='SteelBlue', edgecolor="LightSteelBlue")
plt.show()
sns.boxplot(x=app_train['DAYS_EMPLOYED'])
plt.show()
plt.plot(app_train['DAYS_EMPLOYED'],'*')
plt.show()
#可疑的欄位
#OWN_CAR_AGE Age of client's car
#最大值掉在分配外面
print(app_train['OWN_CAR_AGE'].describe())
plt.hist(app_train['OWN_CAR_AGE'], bins=50, color='SteelBlue', edgecolor="LightSteelBlue")
plt.show()
sns.boxplot(x=app_train['OWN_CAR_AGE'])
plt.show()
plt.plot(app_train['OWN_CAR_AGE'],'*')
plt.show()
#OBS_30_CNT_SOCIAL_CIRCLE How many observation of client's social surroundings with observable 30 DPD (days past due) default
#最大值掉在分配外面
print(app_train['OBS_30_CNT_SOCIAL_CIRCLE'].describe())
plt.hist(app_train['OBS_30_CNT_SOCIAL_CIRCLE'], bins=50, color='SteelBlue', edgecolor="LightSteelBlue")
plt.show()
sns.boxplot(x=app_train['OBS_30_CNT_SOCIAL_CIRCLE'])
plt.show()
plt.plot(app_train['OBS_30_CNT_SOCIAL_CIRCLE'],'*')
plt.show()
```

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
cdf = app_train['AMT_INCOME_TOTAL'].sort_values()
cdf = cdf.reset_index(drop=True)
#plt.plot(list(cdf.index), cdf/cdf.max()) #原始
plt.plot(cdf, (cdf.index/cdf.index.max()) ) # x = 資料 , y = 總數計算每個數據點在所有數據中的位置佔比。

plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
#plt.plot(np.log(list(cdf.index)), cdf/cdf.max()) #原始
plt.plot(np.log(cdf), (cdf.index/cdf.index.max()) ) # x = 資料 , y = 總數計算每個數據點在所有數據中的位置佔比。
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
